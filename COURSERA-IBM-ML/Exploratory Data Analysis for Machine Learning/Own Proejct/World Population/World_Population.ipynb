The data is from [Kaggle](https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset?select=world_population.csv), which tells us about the population in countries all around the world for different years.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "world_population.csv"
data_0 = pd.read_csv(file)
data_0.head()
```

```python
#Listing the data columns
data_0.columns.to_list()
```

```python
#Listing the datatypes of the columns
data_0.dtypes
```

```python
#Describing the data
stats_df_data_0 = data_0.describe()
stats_df_data_0
```

```python
data_1 = data_0.copy()
data_1["Country"] = [str(i) for i in data_0["Country"]]
data_1["Capital"] = [str(i) for i in data_0["Capital"]]
data_1["Continent"] = [str(i) for i in data_0["Continent"]]
data_1["CCA3"] = [str(i) for i in data_0["CCA3"]]
```

```python
from sklearn import preprocessing
```

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 15))
axes0 = ax[0, :]
axes1 = ax[1, :]

axes0[0].scatter(preprocessing.normalize([data_1["Area (km²)"]]), preprocessing.normalize([data_1["2022 Population"]]))
axes0[0].set(xlabel = "Area (km²)", ylabel = "2022 Population", title = "Current population vs Area of the country (km²)")
axes0[1].scatter(preprocessing.normalize([data_1["Area (km²)"]]), preprocessing.normalize([data_1["2022 Population"]]))
axes0[1].set(xlabel = "Density (per km²)", ylabel = "2022 Population", title = "Current population vs Density (per km²)")
axes1[0].scatter(preprocessing.normalize([data_1["Growth Rate"]]), preprocessing.normalize([data_1["2022 Population"]]))
axes1[0].set(xlabel = "Growth Rate", ylabel = "2022 Population", title = "Current population vs Growth Rate")
axes1[1].scatter(preprocessing.normalize([data_1["World Population Percentage"]]), preprocessing.normalize([data_1["2022 Population"]]))
axes1[1].set(xlabel = "World Population Percentage", ylabel = "2022 Population", title = "Current population vs World Population Percentage")
```
